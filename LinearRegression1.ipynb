## Linear Regression 1

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Data x - Acidity of milk, y - density of milk

```python
def readData(filename):
    df = pd.read_csv(filename)
    return df.values
```

```python
x = readData('./Desktop/ML Data/data/linearX.csv')
print(x.shape)
# x is in the form of list so we have to convert into array
x.reshape((99,))

y = readData('./Desktop/ML Data/data/linearY.csv')
print(y.shape)
y.reshape((99,))
x.shape
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# We do Normalization next (mean = 0, standard deviation = 1)
# This improves the efficiency
x = (x - x.mean())/(x.std())
plt.scatter(x, y)
plt.show()
X = x;
Y = y;
```

#### Linear Regression Part

```python
def hypothesis(theta, x):
    return theta[1]*x + theta[0]

def error(X, Y, theta):
    total_error = 0;
    m = X.shape[0];
    for i in range(m):
        total_error += (Y[i] - hypothesis(theta, X[i]))**2
        
    return 0.5*total_error

def gradient(X, Y, theta):
    grad = np.array([0.0, 0.0])
    m = X.shape[0]
    for i in range(m):
        grad[0] += -1*(Y[i]-hypothesis(theta, X[i]))
        grad[1] += -1*(Y[i]-hypothesis(theta, X[i]))*X[i]
    
    return grad

def gradientDescent(X, Y, learning_rate, maxItr):
    grad = np.array([0.0, 0.0])
    theta = np.array([0.0, 0.0])
    
    for i in range(maxItr):
        grad = gradient(X, Y, theta)
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
    
    return theta

theta = gradientDescent(X, Y, learning_rate=0.001, maxItr=1000)
print(theta[0], theta[1])
```

```python
plt.scatter(X, Y)
plt.plot(X, hypothesis(theta, X), color = 'r')
plt.show()
```
